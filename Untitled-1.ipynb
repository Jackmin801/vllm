```python
import pandas as pd

#df_ori = pd.read_parquet("ori100/validation_Qwen--Qwen3-0.6B_flash_bfloat16_1_1xH100_on_Qwen--Qwen3-0.6B.parquet")
#df_full = pd.read_parquet("full100/validation_Qwen--Qwen3-0.6B_flash_bfloat16_1_1xH100_on_Qwen--Qwen3-0.6B.parquet")

df_ori = pd.read_parquet("ori100/validation_deepseek-ai--DeepSeek-R1-Distill-Qwen-1.5B_flash_bfloat16_1_1xH100_on_deepseek-ai--DeepSeek-R1-Distill-Qwen-1.5B.parquet")
df_full = pd.read_parquet("full100/validation_deepseek-ai--DeepSeek-R1-Distill-Qwen-1.5B_flash_bfloat16_1_1xH100_on_deepseek-ai--DeepSeek-R1-Distill-Qwen-1.5B.parquet")
```

```python
import torch
import matplotlib.pyplot as plt

chosen_probs_ori = torch.cat([torch.tensor(i).exp() for i in df_ori['chosen_logprob']])
output_probs_ori = torch.cat([torch.tensor(i).exp() for i in df_ori['output_logprobs']])
chosen_probs_full = torch.cat([torch.tensor(i).exp() for i in df_full['chosen_logprob']])
output_probs_full = torch.cat([torch.tensor(i).exp() for i in df_full['output_logprobs']])

abs_diff_ori = torch.abs(chosen_probs_ori - output_probs_ori)
abs_diff_full = torch.abs(chosen_probs_full - output_probs_full)

fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0,0].scatter(chosen_probs_ori, output_probs_ori, alpha=0.1)
axs[0,0].set_title('BF16: Prefill vs Decode')
axs[0,1].hist(abs_diff_ori, bins=100, alpha=0.5)
axs[0,1].set_title('BF16: Absolute Difference')
axs[1,0].scatter(chosen_probs_full, output_probs_full, alpha=0.1)
axs[1,0].set_title('FP32LM: Prefill vs Decode')
axs[1,1].hist(abs_diff_full, bins=100, alpha=0.5)
axs[1,1].set_title('FP32LM: Absolute Difference')
print(f"BF16 mean: {abs_diff_ori.mean()}, FP32LM mean: {abs_diff_full.mean()}")
from scipy.stats import pearsonr

corr_ori, _ = pearsonr(chosen_probs_ori.numpy(), output_probs_ori.numpy())
corr_full, _ = pearsonr(chosen_probs_full.numpy(), output_probs_full.numpy())
print(f"Pearson correlation (original): {corr_ori:.4f}")
print(f"Pearson correlation (full): {corr_full:.4f}")

plt.show()
```
